```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,df.columns!='price']
y = df.iloc[:,df.columns=='price']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.5)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain.values.ravel())
```

```python
predict_output = model.predict(xtest)
print(predict_output)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(type(ytest), type(predict_output))
print(ytest[:5], predict_output[:5])
```

```python
predict_output = np.argmax(predict_output)
```

```python
predict_output = model.predict(xtest) 
```

```python
print(predict_output.shape)
if len(predict_output.shape) > 1 and predict_output.shape[1] > 1:
    predict_output = np.argmax(predict_output, axis=1)
```

```python
print("ytest unique values:", np.unique(ytest))
print("predict_output unique values:", np.unique(predict_output))
```

```python
ytest = ytest.astype(int)
predict_output = predict_output.astype(int)
```

```python
acc=accuracy_score(ytest,predict_output)
print('The accuracy score for RF:',acc)
```

```python
from sklearn.metrics import r2_score
a=r2_score(ytest,predict_output)
print('The r2 score for RF:',a)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(ytest,predict_output)
print('The mean squared error:',mse)
```
